```python
#ЗАДАНИЕ 5

#Номер проездного билета для проверки
number = 142321

#Разбиваем номер билета на отдельные числа путём взятия остатка от деления на 10 и последующим отбрасыванием последней цифры
number_6 = number % 10
number_5 = (number // 10) % 10
number_4 = (number // 100) % 10
number_3 = (number // 1000) % 10
number_2 = (number // 10000) % 10
number_1 = (number // 100000) % 10

#Если сумма первых трёх цифр совпадает с суммой последних, то билет счастливый
if (number_1 + number_2 + number_3) == (number_4 + number_5 + number_6):
    print('Счастливый билет')
else:
    print('Несчастливый билет')
```

```python
#ЗАДАНИЕ 6

#Запрашиваем тип фигуры
Shape = input('Введите тип фигуры:\n')

#Если пользователь выбрал круг, запрашиваем его радиус
if Shape == 'Круг':
    R = float(input('Введите радиус круга:\n'))
    #Площадь круга - произведение числа пи на радиус в квадрате
    S = round(3.1416 * R ** 2, 2)
    print('Площадь круга:', S)
#Если треугольник – длины трех его сторон
if Shape == 'Треугольник':
    #Запрашиваем длины сторон
    A = float(input('Введите длину стороны A:\n'))
    B = float(input('Введите длину стороны B:\n'))
    C = float(input('Введите длину стороны C:\n'))
    #Вычисляем полупериметр
    P = (A + B + C) / 2
    #Вычисляем площадь по формуле Герона
    S = round((P * (P - A) * (P - B) * (P - C)) ** 0.5, 2)
    print('Площадь треугольника:', S)
#Если прямоугольник – длины двух его сторон
if Shape == 'Прямоугольник':
    #Запрашиваем длины сторон
    A = float(input('Введите длину стороны A:\n'))
    B = float(input('Введите длину стороны B:\n'))
    #Площадь прямоугольника - произведение длин его сторон
    S = round(A * B, 2)
    print('Площадь прямоугольника:', S)
else:
    print('Ошибка')
```
